#### Trabajo de equipo
CompuTech, una empresa de servicios de consulta de AI a través de API, está evaluando cambiar el precio de uno de sus servicios más populares. Tienen como objetivo incrementar 20 USD el ingreso promedio por usuario.


El servicio tiene actualmente un precio de 1 USD por consulta, y la empresa está considerando aumentar el precio a 1.25 USD. Antes de implementar este cambio, desean evaluar cómo este aumento afectará su KPI ventas por usuario. Se llevará a cabo una prueba A/B para respaldar esta decisión.

##### Objetivo del Caso
El objetivo de la prueba es determinar si el cambio de precio tiene un impacto significativo en los siguientes KPIs:

Ingreso promedio por usuario: Ingreso promedio por usuario. Desean que por lo menos aumente un 20 USD.

Tasa de Conversión: Proporción de usuarios que completan la suscripción.


##### Hipótesis
- Hipótesis Nula ($H_0$): El aumento de precio a 1.25 USD no tiene un impacto significativo en la tasa de conversión ni en el ingreso promedio por usuario.
- Hipótesis Alternativa ($H_1$): El aumento de precio a 1.25 USD tiene un impacto significativo en la tasa de conversión y el ingreso promedio por usuario.


#Escenarios de P-Value

1. Escenario 1: p > 0.95

  * Interpretación: No hay evidencia suficiente para rechazar la hipótesis nula. Esto significa que la probabilidad de observar los datos, si la hipótesis nula fuera cierta, es superior al 95%. En otras palabras, es muy probable que la diferencia (o efecto) observada entre los grupos sea debida al azar.

  * Conclusión: En este escenario, generalmente se concluye que no hay suficiente evidencia para afirmar que hay una diferencia significativa entre los grupos.

  * Ejemplo: En tu código, el test de Welch arrojó un p-value mayor que 0.05. Esto indica que no hay suficiente evidencia para rechazar la hipótesis nula de que la media del LTV entre los dos grupos es la misma.

2. Escenario 2: p < 0.05

  * Interpretación: Hay evidencia suficiente para rechazar la hipótesis nula. Esto quiere decir que la probabilidad de observar los datos, si la hipótesis nula fuera cierta, es inferior al 5%. Es poco probable que la diferencia (o efecto) observada sea debida al azar.

  * Conclusión: Se concluye que hay una diferencia significativa entre los grupos, y esta diferencia se considera estadísticamente significativa.

  * Ejemplo: Supongamos que al analizar la tasa de conversión entre los dos grupos en tu código, se obtuvo un p-value de 0.02. Esto indicaría que la diferencia en las tasas de conversión es estadísticamente significativa, y por lo tanto se puede rechazar la hipótesis nula de que las tasas son iguales.

3. Escenario 3: 0.05 < p < 0.95

  * Interpretación: Los resultados no son concluyentes. La probabilidad de observar los datos si la hipótesis nula fuera cierta está en un rango intermedio. No hay suficiente evidencia para rechazar o aceptar la hipótesis nula de forma contundente.

  * Conclusión: Se requiere más investigación, tal vez con una muestra más grande o con una prueba diferente, para obtener resultados más concluyentes.

  * Ejemplo: Imagina que en tu análisis del ARPU, el p-value fuera 0.25. Esto no sería suficiente para rechazar la hipótesis nula de que el aumento del precio tiene un impacto significativo en el ARPU, pero tampoco sería una evidencia contundente para aceptar la hipótesis nula.

```python
# Experimento controlado, distinto número de observaciones
ltv_grupo_a = df_AR['ARPU']  # Más observaciones
ltv_grupo_b = df_BR['ARPU']  # Menos observaciones

# Realizamos un Welch's t-test, que es robusto a tamaños de muestras desiguales
#Sirve para comparar la media de cada grupo individual para determinar si estadísticamente hay una diferencia significante
t_stat, p_value = stats.ttest_ind(ltv_grupo_a, ltv_grupo_b, equal_var=False)

# Mostramos los resultados
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

* Valor p: Representa la probabilidad de observar los resultados obtenidos (o resultados aún más extremos) si no hubiera una diferencia real entre los grupos (es decir, si la hipótesis nula fuera verdadera). Un valor p más pequeño sugiere una mayor evidencia en contra de la hipótesis nula, lo que indica que la diferencia entre los grupos es estadísticamente significativa.

* Valor alto del t-estadístico: Si el número es grande, indica que la diferencia entre las medias es grande en comparación con la variabilidad. Esto sugiere que la diferencia observada es probablemente real y no solo por azar.

* Valor bajo del t-estadístico: Si el número es pequeño, significa que la diferencia entre los promedios no es tan significativa y podría deberse al azar, especialmente si hay mucha variabilidad en los datos.

1. Ingreso promedio por usuario (ARPU)

Para el ARPU, utilizaremos un t-test de Welch para comparar las medias de los dos grupos (precio actual vs. nuevo precio).

```python
# Objectivo: tener diferencia de 20 dolares en el LTV
objective_difference = 20

# Promedio y desviación estandar por grupo
mean_a = df_AR['ARPU'].mean()
mean_b = df_BR['ARPU'].mean()

std_a = np.std(df_AR['ARPU'], ddof=1)
std_b = np.std(df_BR['ARPU'], ddof=1)

n_a = len(df_AR['ARPU'])
n_b = len(df_BR['ARPU'])

# Realizar un t-test de una cola (mayor o igual que el objetivo)
# Null hypothesis H0: (mean_b - mean_a) <= objective_difference
# Alternative hypothesis H1: (mean_b - mean_a) > objective_difference

# Calculate the t-statistic and p-value
pooled_std = np.sqrt(((std_a**2 / n_a) + (std_b**2 / n_b)))
t_stat = (mean_b - mean_a - objective_difference) / pooled_std
p_value = 1 - stats.t.cdf(t_stat, df=min(n_a, n_b) - 1)
#PARA PRUEBA DE COLA IZQUIERDAS QUITAS EL 1 -

# Output the results
print(f"Mean LTV of Group A: {mean_a:.2f}")
print(f"Mean LTV of Group B: {mean_b:.2f}")
print(f"Objective Difference: {objective_difference}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```
